```python
import pandas as pd
df=pd.read_json("E:\Metatalent.ai\Entity Recognition in Resumes.json",lines=True)
df.head(5)
```

```python
df.annotation[81]
```

```python
dic={
    'Name':[],
    'Skills':[],
    'College Name':[],
    'Graduation Year':[], 
    'Degree':[], 
    'Designation':[], 
    'Companies worked at':[], 
    'Email Address':[], 
    'Location':[],
}
```

```python
max=0
for x in range(df.annotation.count()):
    if max<len(df.annotation[x]):
        max=len(df.annotation[x])

for i in range(0,df.annotation.count()):
    Name=" "
    Skills=" "
    College_Name=" "
    Graduation_Year=" " 
    Degree=" "
    Years_of_Experience= " "
    Designation=" "
    Comapanies_worked_at=" "
    EmailAddress=" "
    Location=" "
    for j in range(0,max):
        if len(df.annotation[i])>j and df.annotation[i][j]['label']!=[]:
            if df.annotation[i][j]['label'][0] == "Name":
                Name=df.annotation[i][j]['points'][0]['text']
            elif df.annotation[i][j]['label'][0] == "Skills":
                Skills=Skills+","+df.annotation[i][j]['points'][0]['text']
            elif df.annotation[i][j]['label'][0] == "College Name":
                College_Name=df.annotation[i][j]['points'][0]['text']
            elif df.annotation[i][j]['label'][0] == "Graduation Year":
                Graduation_Year=df.annotation[i][j]['points'][0]['text']
            elif df.annotation[i][j]['label'][0] == "Degree":
                Degree=df.annotation[i][j]['points'][0]['text']
            elif df.annotation[i][j]['label'][0] == "Years of Experience":
                Years_of_Experience=df.annotation[i][j]['points'][0]['text']   
            elif df.annotation[i][j]['label'][0] == "Designation":
                Designation=Designation+","+df.annotation[i][j]['points'][0]['text']
            elif df.annotation[i][j]['label'][0] == "Companies worked at":
                Comapanies_worked_at=Comapanies_worked_at+","+df.annotation[i][j]['points'][0]['text']
            elif df.annotation[i][j]['label'][0] == "Email Address":
                EmailAddress=df.annotation[i][j]['points'][0]['text']
            elif df.annotation[i][j]['label'][0] == "Location":
                Location=df.annotation[i][j]['points'][0]['text']
        else:
            continue
    
    dic['Name'].append(Name)
    dic['Skills'].append(Skills)
    dic['College Name'].append(College_Name)
    dic['Graduation Year'].append(Graduation_Year) 
    dic['Degree'].append(Degree) 
    dic['Designation'].append(Designation) 
    dic['Companies worked at'].append(Comapanies_worked_at)
    dic['Email Address'].append(EmailAddress)
    dic['Location'].append(Location)
    
```

```python
df1=pd.DataFrame(dic)
print(df1.count())
df1.head(10)
```

```python
df1.to_csv("sample_uncleaned_dataset.csv")
```
